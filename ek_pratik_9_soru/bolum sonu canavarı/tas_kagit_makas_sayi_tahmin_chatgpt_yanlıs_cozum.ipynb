```python
import random

def sayi_tahmin_oyunu():
    print("\n1 ile 100 arasında bir sayı giriniz.")
    rastgele_sayi = random.randint(1, 100)
    tahmin_sayisi = 0

    while True:
        tahmin = input()
        if tahmin.lower() == 'esc':
            print("Ana menü")
            return
        if not tahmin.isdigit() or not 1 <= int(tahmin) <= 100:
            print("Yanlış seçim\n1 ile 100 arasında bir sayı gir oyunu oynamak için ya da ana menü için ESC yaz.")
            continue

        tahmin_sayisi += 1
        tahmin = int(tahmin)

        if tahmin < rastgele_sayi:
            print("Daha büyük değerde tahminde bulunun")
        elif tahmin > rastgele_sayi:
            print("Daha küçük değerde tahminde bulunun")
        else:
            print(f"Bingo. {tahmin_sayisi} adet hak kullanarak sayıyı bildiniz: {tahmin}.")
            break

def tas_kagit_makas():
    secenekler = ["Tas", "Kagit", "Makas"]

    while True:
        secim = input("\nSilahini sec: Tas, Kagit, Makas ya da çıkış için ESC yaz: ")
        if secim.lower() == 'esc':
            print("Ana menü")
            return
        if secim.capitalize() not in secenekler:
            print("Yanlış Seçim")
            continue

        bilgisayar_secimi = random.choice(secenekler)
        print(f"{secim}  vs  {bilgisayar_secimi}")
        if secim == bilgisayar_secimi:
            print("Berabere!")
        else:
            kazanan = "Insan" if (secim == "Tas" and bilgisayar_secimi == "Makas") or \
                               (secim == "Kagit" and bilgisayar_secimi == "Tas") or \
                               (secim == "Makas" and bilgisayar_secimi == "Kagit") else "Bilgisayar"
            print(f"{kazanan} kazandı!")

def ana_menu():
    while True:
        print("\n    1: Sayı Tahmin Oyunu,\n    2: Taş, Kağıt, Makas Oyunu,\n    3: Çıkış ")
        secim = input("Seçiminiz : ")

        if secim == '1':
            sayi_tahmin_oyunu()
        elif secim == '2':
            tas_kagit_makas()
        elif secim == '3':
            print("Çıkış yapılıyor...")
            break
        else:
            print("Hatalı seçim")

ana_menu()
```
